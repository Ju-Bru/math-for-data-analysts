Bei der **deskriptiven Statistik** und der **Inferenzstatistik** handelt es sich, neben der explorativen Statistik, um die Hauptbereiche der Statistik.

- Die **deskriptive Statistik** stellt Werkzeuge bereit, um eine **Stichprobe zu beschreiben**.
- Ausgehend von der Stichprobe kann mithilfe der **Inferenzstatistik** eine **Aussage über die Grundgesamtheit** getroffen werden.

## 1. Deskriptive Statistik vs. Inferenzstatistik

Ein Hauptanwendungsgebiet der Statistik ist es, eine Aussage über eine **Grundgesamtheit** zu treffen. Da es in den meisten Fällen aber nicht möglich ist, alle Daten der Grundgesamtheit zu erhalten, wird eine **Stichprobe** gezogen.

Diese Stichprobe kann mit der **deskriptiven Statistik** beschrieben werden, z. B. mit:
- Lageparametern wie dem **Mittelwert**
- Streuungsparametern wie der **Standardabweichung**

Damit kann aber noch **keine Aussage über die Grundgesamtheit** getroffen werden – das ist Aufgabe der **Inferenzstatistik**.

| Merkmal | Deskriptive Statistik | Inferenzstatistik |
|---|---|---|
| Ziel | Stichprobe beschreiben | Schlüsse auf Grundgesamtheit ziehen |
| Ergebnis | Kennzahlen, Tabellen, Grafiken | Hypothesentests, p-Werte |
| Aussage über | Nur die vorliegende Stichprobe | Die gesamte Population |
| Methoden | Mittelwert, Median, Streuung | t-Test, Chi-Quadrat, ANOVA |

## 2. Deskriptive Statistik

Das Ziel der deskriptiven Statistik ist es, Daten umfassend zu **beschreiben**:
- Einen Überblick über ihre Verteilung gewinnen
- Zentrale Kennzahlen berechnen (Mittelwert, Median, Streuung)
- Daten anschaulich darstellen

> **Definition:** Die deskriptive Statistik fasst statistische Methoden zusammen, die zur Beschreibung von Daten anhand von **Kennzahlen**, **Diagrammen**, **Grafiken** und **Tabellen** verwendet werden.

**Wichtig:** In der deskriptiven Statistik werden **keine Schlussfolgerungen** über andere Zeitpunkte oder die Grundgesamtheit gezogen.

### Teilbereiche der deskriptiven Statistik

| Bereich | Beispiele |
|---|---|
| **Lagemaße** | Mittelwert, Median, Modalwert, Summe |
| **Streuungsmaße** | Standardabweichung, Varianz, Spannweite |
| **Häufigkeitstabellen** | Absolute, relative und kumulierte Häufigkeiten |
| **Diagramme** | Histogramme, Balkendiagramme, Boxplots, Streudiagramme |

### 2.1 Beispiel: Deskriptive Statistik in Python

Stichprobe von 10 männlichen Basketballspielern (Körpergröße in Metern):

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Stichprobe: Körpergrößen von 10 Basketballspielern
data = {
    'Spieler': range(1, 11),
    'Körpergröße (m)': [1.62, 1.72, 1.55, 1.70, 1.78, 1.65, 1.64, 1.64, 1.66, 1.74]
}

df = pd.DataFrame(data)
print(df.to_string(index=False))
```

```python
größe = df['Körpergröße (m)']

print("Deskriptive Statistiken")
print(f"Anzahl (N):            {len(größe)}")
print(f"Mittelwert:            {größe.mean():.4f} m")
print(f"Median:                {größe.median():.4f} m")
print(f"Modalwert:             {größe.mode()[0]:.4f} m")
print(f"Standardabweichung:    {größe.std():.4f} m")
print(f"Varianz:               {größe.var():.4f} m²")
print(f"Minimum:               {größe.min():.4f} m")
print(f"Maximum:               {größe.max():.4f} m")
print(f"Spannweite:            {größe.max() - größe.min():.4f} m")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Histogramm
axes[0].hist(größe, bins=5, color='steelblue', edgecolor='white', alpha=0.8)
axes[0].axvline(größe.mean(), color='red', linestyle='--', label=f'Mittelwert: {größe.mean():.2f} m')
axes[0].axvline(größe.median(), color='orange', linestyle='--', label=f'Median: {größe.median():.2f} m')
axes[0].set_title('Histogramm: Körpergröße')
axes[0].set_xlabel('Körpergröße (m)')
axes[0].set_ylabel('Häufigkeit')
axes[0].legend()

# Boxplot
axes[1].boxplot(größe, vert=True, patch_artist=True,
                boxprops=dict(facecolor='steelblue', alpha=0.6))
axes[1].set_title('Boxplot: Körpergröße')
axes[1].set_ylabel('Körpergröße (m)')
axes[1].set_xticklabels(['Basketballspieler'])

plt.tight_layout()
plt.show()
```

## 3. Inferenzstatistik

Im Gegensatz zur deskriptiven Statistik zielt die **Inferenzstatistik** darauf ab, Aussagen über die **Grundgesamtheit** zu treffen.

> **Definition:** Die Inferenzstatistik ist ein Teilgebiet der Statistik, das verschiedene Analyseverfahren nutzt, um aus Stichprobendaten **Rückschlüsse auf die Grundgesamtheit** zu ziehen.

Da es nahezu unmöglich ist, die gesamte Population zu befragen, wird eine **Stichprobe** verwendet. Da die Aussage auf Grundlage einer Stichprobe getroffen wird, kann sie immer nur mit einer bestimmten **Wahrscheinlichkeit** bestätigt oder widerlegt werden.

### Verfahren der Inferenzstatistik

| Fragestellung | Verfahren |
|---|---|
| Gruppenunterschiede (2 Gruppen) | t-Test |
| Gruppenunterschiede (3+ Gruppen) | Varianzanalyse (ANOVA) |
| Kategoriale Variablen | Chi-Quadrat-Test |
| Zusammenhang zwischen Variablen | Korrelationsanalyse |
| Vorhersage | Regressionsanalyse |

### 3.1 Beispiel: Einstichproben t-Test

**Fragestellung:** Sind Basketballspieler im Durchschnitt größer als die männliche Bevölkerung insgesamt?

- Bekannte durchschnittliche Körpergröße deutscher Männer (Grundgesamtheit): **1,74 m**
- **H₀ (Nullhypothese):** Basketballspieler sind nicht größer als der Durchschnitt (μ = 1,74 m)
- **H₁ (Alternativhypothese):** Basketballspieler sind größer (μ > 1,74 m)

```python
from scipy import stats

# Bekannter Populationsmittelwert (deutsche Männer)
mu_population = 1.74

# Einstichproben t-Test
t_stat, p_value = stats.ttest_1samp(größe, popmean=mu_population)

print("Einstichproben t-Test")
print(f"Mittelwert Stichprobe: {größe.mean():.4f} m")
print(f"Mittelwert Population: {mu_population:.4f} m")
print(f"t-Statistik:           {t_stat:.4f}")
print(f"p-Wert (zweiseitig):   {p_value:.4f}")
print(f"p-Wert (einseitig):    {p_value/2:.4f}")
print()

alpha = 0.05
if p_value / 2 < alpha and t_stat > 0:
    print(f"Ergebnis: p < {alpha} → H₀ wird abgelehnt.")
    print("→ Basketballspieler sind signifikant größer als der Bevölkerungsdurchschnitt.")
else:
    print(f"Ergebnis: p ≥ {alpha} → H₀ wird nicht abgelehnt.")
    print("→ Kein signifikanter Unterschied zur Bevölkerung nachweisbar.")
```

### 3.2 Beispiel: Unabhängiger t-Test

**Fragestellung:** Sind Basketballspieler größer als Fußballspieler?

- **H₀:** Es gibt keinen Größenunterschied zwischen Basketball- und Fußballspielern
- **H₁:** Basketballspieler sind größer als Fußballspieler

```python
# Stichprobe Fußballspieler (simulierte Daten)
np.random.seed(42)
fussball = pd.Series([1.70, 1.68, 1.75, 1.72, 1.69, 1.71, 1.67, 1.73, 1.66, 1.74])

print("Gruppenvergleich")
print(f"Basketballspieler: M = {größe.mean():.3f} m, SD = {größe.std():.3f} m")
print(f"Fußballspieler:    M = {fussball.mean():.3f} m, SD = {fussball.std():.3f} m")
print()

# Unabhängiger t-Test
t_stat2, p_value2 = stats.ttest_ind(größe, fussball)

print("Unabhängiger t-Test")
print(f"t-Statistik: {t_stat2:.4f}")
print(f"p-Wert:      {p_value2:.4f}")
print()

if p_value2 < alpha:
    print(f"Ergebnis: p < {alpha} → Signifikanter Unterschied!")
else:
    print(f"Ergebnis: p ≥ {alpha} → Kein signifikanter Unterschied.")

# Visualisierung
fig, ax = plt.subplots(figsize=(8, 5))
ax.boxplot([größe, fussball], labels=['Basketball', 'Fußball'],
           patch_artist=True,
           boxprops=dict(alpha=0.6),
           medianprops=dict(color='black', linewidth=2))
ax.set_title('Körpergröße: Basketball vs. Fußball')
ax.set_ylabel('Körpergröße (m)')
plt.tight_layout()
plt.show()
```

## 4. Zusammenfassung

| | Deskriptive Statistik | Inferenzstatistik |
|---|---|---|
| **Auch bekannt als** | Beschreibende Statistik | Schließende Statistik |
| **Ziel** | Daten beschreiben | Auf Grundgesamtheit schließen |
| **Bezieht sich auf** | Stichprobe | Grundgesamtheit (Population) |
| **Typische Kennzahlen** | Mittelwert, Median, SD | t-Wert, p-Wert, Konfidenzintervall |
| **Typische Methoden** | Häufigkeitstabellen, Diagramme | t-Test, ANOVA, Chi-Quadrat, Regression |
| **Schlussfolgerungen** | Nur für vorliegende Daten | Über die gesamte Population |
| **Unsicherheit** | Keine (exakte Beschreibung) | Immer (Irrtumswahrscheinlichkeit) |

---

> Die **deskriptive Statistik** ist die Grundlage und Vorarbeit für die **Inferenzstatistik**.  
> Beide Bereiche ergänzen sich und sind in der wissenschaftlichen Datenanalyse unverzichtbar.
